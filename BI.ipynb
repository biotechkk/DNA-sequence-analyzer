```python
# Install seaborn if not already installed
!pip install seaborn pandas matplotlib --quiet

# ---- 1. Import libraries ----
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ---- 2. Load your dataset ----
# Replace 'gene_expression.csv' with your file path
# The CSV should have genes as rows and samples/conditions as columns
from google.colab import files
uploaded = files.upload()  # Upload your file

df = pd.read_csv(next(iter(uploaded)))

print("Data preview:")
print(df.head())

# ---- 3. Optional: set index to gene names ----
# Assuming first column is 'Gene'
if df.columns[0].lower() in ["gene", "gene_id", "gene_name"]:
    df.set_index(df.columns[0], inplace=True)

# ---- 4. Create the heatmap ----
plt.figure(figsize=(10, 8))
sns.heatmap(df, cmap="viridis", annot=False)
plt.title("Gene Expression Heatmap", fontsize=16)
plt.xlabel("Samples / Conditions")
plt.ylabel("Genes")
plt.tight_layout()
plt.show()
```
